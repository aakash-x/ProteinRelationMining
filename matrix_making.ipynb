```python
import pandas as pd
import numpy as np
import nltk
from nltk.corpus import treebank
##################################################################
df1 = pd.read_csv("input.csv")
pname=df1['PROTEIN_NAME'].drop_duplicates().values.tolist()
#def function1(protein_list):
def function1():
    "Searching all the proteins"
    df1 = pd.read_csv("input.csv")
    print('Loaded')
    #print(df1)
    protein_list = df1['PROTEIN_NAME'].drop_duplicates().values.tolist()
    #print(protein_list)
    length=(len(protein_list))
    #print(length)
    Matrix = [[0 for x in range(length)] for y in range(length)]
    for protein in protein_list:
        spe_protein = df1.loc[df1['PROTEIN_NAME'] == protein] #Selecting dataframe of specific Protein
        a = spe_protein.ABSTRACT.str.cat(sep=', ')            #Combining Column of all the protein
        #print(a)
        #print(protein)
        tokens = nltk.word_tokenize(a)
        tagged = nltk.pos_tag(tokens)
        #print(tokens)
        for i in tagged:
            if(i[1]=='NNP') and (i[0] in protein_list):
                #print(i[0],end=' ')
                #print(i[0] in protein_list)
                row = protein_list.index(protein)
                col = protein_list.index(i[0])
                #row = protein and col = frequent
                Matrix[row][col]+=1
                
    #print(Matrix)
    
    
    print('    ', end='', flush=True)
    print(pname)
    t=0
    print('')
    for x in Matrix:
        print(pname[t]+"    ", end='', flush=True)
        print(*x, sep="       ")
        t+=1
        print('')
    

                        
##################################################################
function1()
```
