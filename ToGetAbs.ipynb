```python
from pubmed_lookup import PubMedLookup
import pandas as pd
email = 'rakesh161@gmail.com'
km=pd.read_csv("Protein_IDs.csv")
k=km.Protein.head(1000)
abst=[]
year=[]
print("Finding Abstracts")
count=1;
for i in k:
    try:
        url = 'http://www.ncbi.nlm.nih.gov/pubmed/'+str(i)
        lookup = PubMedLookup(url, email)
        from pubmed_lookup import Publication
        publication = Publication(lookup,resolve_doi=False)    # Use 'resolve_doi=False' to keep DOI UR
        print(count)
        count+=1
        print("ID :"+str(i))
        print("Abstract :"+repr(publication.abstract))
        abst.append(repr(publication.abstract))
        year.append(publication.year)
    except:
        print("Unable to get Abstract of PMID "+str(i))
        abst.append("null")
        year.append("null")
d = {'PMID': k,'Year':year, 'Abstract': abst}
df = pd.DataFrame(data=d)
df.to_csv('CSV_1000.csv', sep='\t',index=False)
print("Done")
```
